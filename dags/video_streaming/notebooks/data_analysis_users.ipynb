### Outliers

Para este dataframe se eligió separar las outliers del datframe "users.csv" y crear otro con todas esas outliers y otro sin outliers. Esto con el fin de no eliminarlas ya que pueden aportar un insight bueno de como se comporta esa parte de la muestra

### Calcular Outliers con la columna "total_watch_time_hours"

```python

# Calcular Q1 (25%) y Q3 (75%)
Q1 = df["total_watch_time_hours"].quantile(0.25)
Q3 = df["total_watch_time_hours"].quantile(0.75)
IQR = Q3 - Q1

# Definir los límites para detectar outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Límite inferior:", lower_bound)
print("Límite superior:", upper_bound)

# Filtrar los outliers
outliers = df[(df["total_watch_time_hours"] < lower_bound) |
              (df["total_watch_time_hours"] > upper_bound)]

print("Cantidad de outliers:", outliers.shape[0])
print(outliers[["user_id", "age", "total_watch_time_hours"]].head(12))
```

### Separacion de dataframe sin outliers y solo outliers

```python
# DataFrame con outliers
df_outliers_users = df[(df["total_watch_time_hours"] < lower_bound) |
                 (df["total_watch_time_hours"] > upper_bound)].copy()

# DataFrame sin outliers (datos normales)
df_normal_users = df[(df["total_watch_time_hours"] >= lower_bound) &
               (df["total_watch_time_hours"] <= upper_bound)].copy()

print("Filas totales:", df.shape[0])
print("Normales:", df_normal_users.shape[0])
print("Outliers:", df_outliers_users.shape[0])
```

```python
# ===== 2.2 Inferential Analysis =====
from IPython.display import Markdown, display
display(Markdown("## 2.2 Inferential Analysis\nTrabajando con `df_users_normal` (sin outliers)."))

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Estadística / ML
from scipy import stats
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, classification_report, roc_auc_score, roc_curve
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression

# Series de tiempo
from statsmodels.tsa.seasonal import seasonal_decompose

# Si df_users_normal no está en memoria, lo cargamos desde CSV:
try:
    df = df_users_normal.copy()
except NameError:
    df = pd.read_csv('df_users_normal.csv')

print("Shape:", df.shape)
df.head(3)
```

```python
display(Markdown("### 2.2.1 Hypothesis Testing — Premium vs Basic user engagement"))

metric = 'total_watch_time_hours'  # engagement
group_col = 'subscription_type'    # debe contener 'Premium' y 'Basic'

# Filtrar y dejar solo Premium/Basic
sub = df[df[group_col].isin(['Premium','Basic'])][[group_col, metric]].dropna()
g1 = sub[sub[group_col]=='Premium'][metric].values
g2 = sub[sub[group_col]=='Basic'][metric].values

print(f"N Premium={len(g1)}, N Basic={len(g2)}")

# Normalidad (Shapiro para n<=5000; para n grande, aplicamos regla del tamaño y seguimos con t-test robusto)
def is_normal(x):
    x = x[~np.isnan(x)]
    if len(x) < 3:
        return False
    n = min(len(x), 5000)
    stat, p = stats.shapiro(x[:n])
    return p > 0.05

normal_g1, normal_g2 = is_normal(g1), is_normal(g2)

# Homogeneidad de varianzas
lev_stat, lev_p = stats.levene(g1, g2)
equal_var = lev_p > 0.05

if normal_g1 and normal_g2:
    t_stat, p_val = stats.ttest_ind(g1, g2, equal_var=equal_var)
    test_used = f"t-test independiente (equal_var={equal_var})"
    # Tamaño de efecto (Cohen's d)
    pooled_sd = np.sqrt(((len(g1)-1)*np.var(g1, ddof=1) + (len(g2)-1)*np.var(g2, ddof=1)) / (len(g1)+len(g2)-2))
    effect = (np.mean(g1) - np.mean(g2)) / pooled_sd
else:
    t_stat, p_val = stats.mannwhitneyu(g1, g2, alternative='two-sided')
    test_used = "Mann-Whitney U (no normal)"
    # Tamaño de efecto aproximado (rank-biserial)
    n1, n2 = len(g1), len(g2)
    U = t_stat
    effect = 1 - (2*U)/(n1*n2)

print(f"Prueba usada: {test_used}")
print(f"Estadístico={t_stat:.4f}  p-valor={p_val:.6f}")
print(f"Tamaño de efecto (Cohen d o rank-biserial): {effect:.3f}")
print(f"Media Premium={np.mean(g1):.2f}  Media Basic={np.mean(g2):.2f}")
```

```python
display(Markdown("### 2.2.2 Correlation Analysis — Age, viewing time, completion rates"))

# Intenta detectar posibles columnas
possible_cols = ['age','total_watch_time_hours','completion_rate','sessions_count','movies_completed','series_completed']
cols = [c for c in possible_cols if c in df.columns]

if len(cols) < 2:
    raise ValueError(f"Necesitas al menos 2 columnas numéricas de {possible_cols}. Encontradas: {cols}")

num = df[cols].select_dtypes(include=[np.number]).dropna()
corr = num.corr(method='spearman')

# p-values para Spearman
pvals = pd.DataFrame(np.ones_like(corr), index=corr.index, columns=corr.columns)
for i in corr.columns:
    for j in corr.columns:
        if i == j:
            pvals.loc[i,j] = 0.0
        else:
            r, p = stats.spearmanr(num[i], num[j])
            pvals.loc[i,j] = p

display(Markdown("**Matriz de correlaciones (Spearman):**"))
display(corr.round(3))
display(Markdown("**p-values de las correlaciones:**"))
display(pvals.round(4))

plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation heatmap (Spearman)")
plt.tight_layout()
plt.show()
```

```python
display(Markdown("### 2.2.3 Clustering — User segmentation (K-means)"))

features_try = ['age','total_watch_time_hours','completion_rate','sessions_count']
feat = [f for f in features_try if f in df.columns]
if len(feat) < 2:
    raise ValueError(f"Para clustering necesitas >=2 features. Revisa que existan entre: {features_try}")

X = df[feat].dropna().copy()

scaler = StandardScaler()
Xs = scaler.fit_transform(X)

best_k, best_score = None, -1
for k in range(2, 7):
    km = KMeans(n_clusters=k, n_init=10, random_state=42)
    labels = km.fit_predict(Xs)
    score = silhouette_score(Xs, labels)
    if score > best_score:
        best_k, best_score = k, score

km = KMeans(n_clusters=best_k, n_init=10, random_state=42)
labels = km.fit_predict(Xs)
X = X.copy()
X['cluster'] = labels
display(Markdown(f"**k óptimo = {best_k} (silhouette={best_score:.3f})**"))

# Perfilado de clusters
profile = X.groupby('cluster').mean().round(2)
display(Markdown("**Perfil de clusters (promedios por feature):**"))
display(profile)

# Visualización 2D con PCA
pca = PCA(n_components=2, random_state=42)
coords = pca.fit_transform(Xs)
plt.figure(figsize=(6,5))
sns.scatterplot(x=coords[:,0], y=coords[:,1], hue=labels, palette="Set2", s=40)
plt.title("Clusters (PCA 2D)")
plt.xlabel("PC1"); plt.ylabel("PC2")
plt.tight_layout()
plt.show()
```

```python
display(Markdown("### 2.2.4 Regression — Predict user retention"))

# 1) Intentar encontrar un label real
label_candidates = ['retained_next_month','is_active_next_month','retention_label','churn','is_retained']
label_col = next((c for c in label_candidates if c in df.columns), None)

if label_col is None:
    # Proxy de retención (ajusta el umbral si tienes otra métrica)
    thr = df['total_watch_time_hours'].median()
    df['retained_proxy'] = (df['total_watch_time_hours'] >= thr).astype(int)
    label_col = 'retained_proxy'
    print(f"**No se encontró label real. Usando proxy `{label_col}` basado en mediana de watch time ({thr:.2f}).**")

# Features para el modelo
X_cols = [c for c in ['age','total_watch_time_hours','completion_rate','sessions_count'] if c in df.columns]
X = df[X_cols].fillna(df[X_cols].median())
y = df[label_col].astype(int)

# Split & scale
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s  = scaler.transform(X_test)

# Modelo
clf = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
clf.fit(X_train_s, y_train)
y_pred = clf.predict(X_test_s)
y_proba = clf.predict_proba(X_test_s)[:,1]

print("=== Classification report ===")
print(classification_report(y_test, y_pred, digits=3))

try:
    auc = roc_auc_score(y_test, y_proba)
    print(f"ROC-AUC: {auc:.3f}")
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    plt.figure(figsize=(5,4))
    plt.plot(fpr, tpr, label=f"AUC={auc:.3f}")
    plt.plot([0,1],[0,1],'--',color='grey')
    plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
    plt.title("ROC curve — Retention model")
    plt.legend()
    plt.tight_layout()
    plt.show()
except Exception as e:
    print("No se pudo calcular ROC-AUC:", e)

# Importancias (coeficientes)
coefs = pd.Series(clf.coef_[0], index=X_cols).sort_values(ascending=False)
display(Markdown("**Importancia de variables (coeficientes del modelo):**"))
display(coefs.to_frame('coef').round(3))
```

```python
display(Markdown("### 2.2.5 Time Series — Seasonal viewing trends"))

date_col = 'registration_date'  # ahora usamos la columna real del CSV

ts = df[[date_col,'total_watch_time_hours']].dropna().copy()
ts[date_col] = pd.to_datetime(ts[date_col], errors='coerce')
ts = ts.dropna(subset=[date_col]).set_index(date_col).sort_index()

# Serie mensual (suma de horas vistas por mes)
m = ts['total_watch_time_hours'].resample('M').sum()
display(Markdown("**Serie mensual de horas vistas (suma):**"))
display(m.tail())

plt.figure(figsize=(8,3))
m.plot()
plt.title("Viewing time — Monthly sum")
plt.ylabel("hours")
plt.tight_layout()
plt.show()

if len(m) >= 24:
    from statsmodels.tsa.seasonal import seasonal_decompose
    decomp = seasonal_decompose(m, model='additive', period=12)
    decomp.plot()
    plt.suptitle("Seasonal decomposition (additive, period=12)", y=1.02)
    plt.tight_layout()
    plt.show()
else:
    print("No hay suficientes puntos (>=24) para descomponer estacionalidad. Se muestra la tendencia global.")
```
